```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def gauss(x):
    y = (1 / np.sqrt(2 * math.pi)) * np.exp(-1 * (x**2 / 2))
    return y
```

```python
x = np.arange(-10, 10, 0.01)
y = gauss(x)
```

```python
fig = plt.figure()
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("K(x)")
fig.savefig("./fig/gauss.png")
```

```python
data = np.array([-4, 1, 2])
def density(x, h):
    n = len(data)
    k = np.sum(gauss((x - data) / h))
    y = 1 / (n*h) * k
    return y
```

```python
x = np.arange(-10, 10, 0.01)
y = []
h = 1
for value in x:
    y.append(density(value, h))
```

```python
fig = plt.figure()
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
fig.savefig("./fig/density-h1.png")
```
